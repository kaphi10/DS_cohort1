#### Creating custom index in series

```python
index_name=['Rice','Beans','Yam','Potatoes']
number=[300,400,500,600]
series2=pd.Series(number, index=index_name)
series2
```

Performing mathematical operation on two numerical pandas series object, They must have thesame index

```python
# Create another series3
amount=[200,800,700,100]
series3=pd.Series(amount, index=['Rice','Beans','Yam','Potatoes'])
series3
```

```python
total_am=series2+series3
total_am
```

### Creating DataFrame 

```python
#Using Dictionary of list
item1={'Product_name':['Rice','Beans','Yam','Potatoes'],'Product_price':[100,200,300,400]}
df=pd.DataFrame(item1)
df
```

```python
item1={'Product_name':['Rice','Beans','Yam','Potatoes'],'Product_price':[100,200,300,400]}
df=pd.DataFrame(item1,index=['R','B','Y','P'])
df
```

```python
#using List Of Dictionary
item=[{'Product_name':'Yam','Product_price':200},
     {'Product_name':'Rice','Product_price':100},
     {'Product_name':'Tomatoes','Product_price':1000}]
df1=pd.DataFrame(item)
df1
```

```python
# Using Series to create DataFrame
product=pd.Series(['Iphone','Geonee','Samsung','Tecno'])
amount=pd.Series([10000,500,700,100])
data={'Product_name':product,'Product_price':amount}
df=pd.DataFrame(data)
df
```

```python
#using list to create DataFrame
item_list=[['Iphone',10000],['Geonee',500],['Samsung',700]]
df=pd.DataFrame(item_list, columns=['Product_name','Price'])
df
```

```python
#pro
item1={'Product_name':['Rice','Beans','Yam','Potatoes'],'Product_price':[100,200,300,400]}
index_=['R','B','Y','P']
df=pd.DataFrame(item1,index=index_)
df
```
