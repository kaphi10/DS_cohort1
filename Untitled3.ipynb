```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline
```

```python
df=sb.load_dataset('iris')
df.head()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df['species'].unique()
```

```python
plt.figure(figsize=[6,8])
plt.subplot(2,1,2)
plt.hist(df['sepal_length'], bins=20)
plt.title('Distribution of Sepal length')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')

plt.subplot(2,1,1)
plt.scatter(x=df['sepal_length'],y=df['sepal_width'])
plt.title('Relationship between Sepal length and width')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal width')
plt.tight_layout()
plt.show()
```

```python
fig, ax=plt.subplots(nrows=2,figsize=[6,8])
ax[0].hist(df['petal_length'], bins=20)
ax[0].set_title('Petal length distribution')
ax[0].set_ylabel('Frequency')
ax[0].set_xlabel('petal length')

ax[1].bar(df['species'].value_counts().index, df['species'].value_counts().values)
ax[1].set_title('Specie Counts')
ax[1].set_xlabel('Species')
ax[1].set_ylabel('Counts')
plt.tight_layout()
plt.show()
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=[8,4])
sb.barplot(data=df, x='species',y='petal_length', ax=axes[0])


sb.boxplot(data=df,x='species',y='petal_length', ax=axes[1])
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2,figsize=[8,6])
sb.barplot(data=df, x='species',y='petal_length', ax=axes[0,0])
axes[0,0].set_title('total petal length by  Species')

sb.boxplot(data=df,x='species',y='petal_length', ax=axes[0,1])
axes[0,1].set_title('Distribution of Species by petal length')

sb.scatterplot(data=df, x='petal_length', y='petal_width', hue='species', ax=axes[1,0])
axes[1,0].set_title('Species Petal length vs Width ')
axes[1,0].legend(loc='upper left')

sb.violinplot(data=df, x='species', y='petal_width',ax=axes[1,1])
axes[1,1].set_title('Distribution of Species by petal width')
plt.tight_layout()
```
